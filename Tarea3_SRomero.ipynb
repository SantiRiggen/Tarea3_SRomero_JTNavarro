### Ejercicio 1:
>1. Resolver la siguiente ecuación recursiva usando funciones como se vió en clase
$$ D_{n}=(n-1) D_{n-1}+(n-1) D_{n-2} \quad n\ge 3$$
con $D_1=0$ y $D_2 = 1$

Código de solución estudiante 1

```python
# Código de solución Santiago Romero
import numpy as np

def recursiva(n:'Cantidad elementos'):
    x = np.zeros(n-1)
    D1 = 0
    D2 = 1
    x[0] = D1
    x[1] = D2
    
    def vector(i):
        nonlocal x
        x[i-1] = (i-1) * x[i-2] + (i-1) * x[i-3]
    [vector(i) for i in range(3, n)]
    return x
recursiva(6)
```

### Ejercicio 6
>6. Use a nested list comprehension to find all of the numbers from 1-1000 that are divisible by any single digit besides 1 (2-9). 

```python
# Código de solución Santiago Romero
list1 = [i for i in np.arange(1,1000) if 
         [num for num in range(2,9) if
         i % num == 0]]
list1
```
